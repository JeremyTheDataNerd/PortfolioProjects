```python
import pandas as pd
```

## What percentage of all licensed sets ever released were Star Wars themed?

```python
df = pd.read_csv('Dataset/lego_sets.csv')
df.head()
```

```python
parent_themes = pd.read_csv('Dataset/parent_themes.csv')
sets = pd.read_csv('Dataset/sets.csv')
themes = pd.read_csv('Dataset/themes.csv')
parent_themes.head()
```

```python
df_merged = pd.merge(right= parent_themes, left= df, right_on=['name'], left_on=['parent_theme'], how='inner')
df_merged
```

```python
df_merged = df_merged.drop(columns= 'name_y')
df_merged
```

```python
# Check duplicates on the 'set_num' column
df['set_num'].nunique()
```

```python
df[df['set_num'].isna()]
```

```python
licensed = df_merged[df_merged['is_licensed'] == True]
licensed = licensed.dropna(subset= ['set_num'])
licensed
```

```python
star_wars = licensed[licensed['parent_theme'] == 'Star Wars']
star_wars.head(25)
```

```python
the_force = int((star_wars.shape[0]/licensed.shape[0])*100)
print(str(the_force) + '%')
```

```python
# Based on the above, 51% of licensed sets are Star Wars themed
```

## Which year was Star Wars not the most popular licensed theme?

```python
licensed_sorted = licensed.sort_values('year')
licensed_sorted['count'] = 1
sum_df = licensed_sorted.groupby(['year','parent_theme']).sum().reset_index()
sum_df
```

```python
max_df = sum_df.sort_values('count', ascending= False).drop_duplicates('year')
max_df
```

```python
max_df = max_df.sort_values('year')
max_df
```
