```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
plt.style.use('ggplot')
import seaborn as sns
import os

mpl.rcParams['figure.figsize'] = (12, 8)
```

```python
pwd = os.getcwd()
pwd
```

```python
filepath = pwd + '/movies.csv'
```

```python
df_import = pd.read_csv(filepath)
df_import.head()
```

```python
df_modified = df_import.copy()
df_modified.head()
```

```python
# To check each column if there is any missing value
df_modified.isna().sum()
```

```python
# Determine the "Percentage of Missing Values" for each column

for col in df_modified.columns:
    pct_missing = np.mean(df_modified[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# Drop the rows which contain the missing values

df_modified = df_modified.dropna()
df_modified.isna().sum()
```

```python
# Check data types for our columns
df_modified.dtypes
```

```python
# Change data type for columns
df_modified['budget'] = df_modified['budget'].astype('int64')
df_modified['gross'] = df_modified['gross'].astype('int64')
```

```python
df_modified.head()
```

```python
# Create correct year column
df_modified['yearcorrect'] = df_modified['released'].str.extract(pat='([0-9]{4})').astype(str)
df_modified
```

```python
df_modified = df_modified.sort_values(by=['gross'], ascending = False)
```

```python
# Drop any duplicates
df_modified['company'].drop_duplicates().sort_values(ascending=False)
```

```python
# Hypothesis: Budget & Company have high correlation to the gross revenue
# Scatter plot with budget vs gross 

plt.scatter(x=df_modified['budget'], y=df_modified['gross'])
plt.title('Budget vs Gross Revenue', fontsize = 16, fontweight = 'bold')
plt.xlabel('Budget for Film', fontsize = 13)
plt.ylabel('Gross Revenue', fontsize =13)
plt.show()
```

```python
df_modified.head()
```

```python
# Plot budget vs gross using Seaborn
sns.set_style()
sns.regplot(x='budget', y='gross', data=df_modified, scatter_kws={'color':'orange'}, line_kws={'color':"red"})
```

```python
# Take a look at the correlation between numeric fields
df_modified.corr(method='pearson') #pearson, kendall, spearman
```

```python
correlation_matrix = df_modified.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features', fontsize = 16)
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_numerized = df_modified.copy()

for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
df_numerized
```

```python
df_modified
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features', fontsize = 16)
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
correlation_mat = df_numerized.corr()
corr_pairs = correlation_mat.unstack()
corr_pairs
```

```python
corr_pairs_sorted = corr_pairs.sort_values()
corr_pairs_sorted
```

```python
high_corr = corr_pairs_sorted[corr_pairs_sorted > 0.5]
high_corr
# Based on this, we can conclude the votes and budget have the highest correlation to gross revenue
```
